```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def g(N, mu, sigma=1.):
    return sigma * (mu + np.random.randn(N, 2))
```

```python
mu = np.array([0,0])
mu2 = np.array([2.5,2.5])
x = g(200, mu)
x2 = g(200, mu2)
```

```python
plt.scatter(x[:,0], x[:,1], color="blue")
plt.scatter(x2[:,0], x2[:,1], color="red")
```

```python
x_train = np.concatenate((x[:180], x2[:180]))
x_test = np.concatenate((x[180:], x2[180:]))
```

```python
y_train = np.array([0] * 180 + [1] * 180)
y_test = np.array([0] * 20 + [1] * 20)
```

```python
np.save("data/x_train_2d.npy", x_train)
np.save("data/x_test_2d.npy", x_test)
np.save("data/y_train_2d.npy", y_train)
np.save("data/y_test_2d.npy", y_test)
```
