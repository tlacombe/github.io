```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def g(N, mu, sigma=1.):
    return sigma * (mu + np.random.randn(N, 2))
```

```python
mu = np.array([0,0])
mu2 = np.array([2.5,2.5])
x = g(200, mu)
x2 = g(200, mu2)
```

```python
plt.scatter(x[:,0], x[:,1], color="blue")
plt.scatter(x2[:,0], x2[:,1], color="red")
```

```python
x_train = np.concatenate((x[:180], x2[:180]))
x_test = np.concatenate((x[180:], x2[180:]))
```

```python
y_train = np.array([0] * 180 + [1] * 180)
y_test = np.array([0] * 20 + [1] * 20)
```

# Code for the kNN classifier

```python
N1, N2, N3 = 300, 200, 200  # total nb points
n1, n2, n3 = 100, 100, 100   # train set sizes

mu = np.array([0,0])
mu2 = np.array([3,3])
mu3 = np.array([-2,4])
x = g(N1, mu, sigma=2.)
x2 = g(N2, mu2, sigma=0.5)
x3 = g(N3, mu3)
plt.scatter(x[:,0], x[:,1], color="blue", marker='x')
plt.scatter(x2[:,0], x2[:,1], color="red", marker='x')
plt.scatter(x3[:,0], x3[:,1], color="green", marker='x')

x_train = np.concatenate((x[:n1], x2[:n2], x3[:n3]))
x_test =  np.concatenate((x[n1:], x2[n2:], x3[n3:]))
y_train = np.array([0] * n1 + [1] * n2 + [2] * n3)
y_test = np.array([0] * (N1 - n1) + [1] * (N2 - n2) + [2] * (N3 - n3))

np.save("data/x_train_NN.npy", x_train)
np.save("data/x_test_NN.npy", x_test)
np.save("data/y_train_NN.npy", y_train)
np.save("data/y_test_NN.npy", y_test)
```

# Generation of polynomial reg data

```python
N = 100
# size of the training set
n = 50

theta = np.array([1, -2, 3.14, 1, -1, 1])

def poly(x, theta):
    P = np.array([np.sum([xi**p * theta[p] for p in range(len(theta))]) for xi in x])
    return P

x = 2 * np.random.rand(N) - 1

y = poly(x, theta) + 0.3 * np.random.randn(len(x))

x_train = x[:n]
x_test = x[n:]
y_train = y[:n]
y_test = y[n:]

np.save("data/x_train_linreg.npy", x_train)
np.save("data/x_test_linreg.npy", x_test)
np.save("data/y_train_linreg.npy", y_train)
np.save("data/y_test_linreg.npy", y_test)
```
